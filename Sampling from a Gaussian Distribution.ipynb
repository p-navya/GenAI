# Python Implementation: Sampling from a Gaussian Distribution


## Definition
The Gaussian distribution, also known as the normal distribution, is a continuous probability distribution characterized by a symmetric bell-shaped curve. It describes how the values of a variable are distributed around the mean, with most values clustering around the mean and fewer values appearing as they move away.


It is widely used in statistics and machine learning because many natural phenomena 
(e.g., heights, measurement errors, noise) tend to follow a normal distribution.

## Mathematical Formulation 

### Probability Density Function (PDF) of a Gaussian Distribution

The probability density function (PDF) of a Gaussian (normal) distribution is given by:

$$
P(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left(-\frac{(x - \mu)^2}{2 \sigma^2}\right)
$$

Where:

- $\mu$ is the **mean** (central value)  
- $\sigma$ is the **standard deviation** (spread of the data)  
- $x$ is a **continuous variable**

---

### Latent Variable Models and Sampling

In **latent variable generative models**, such as Variational Autoencoders (VAEs), a common assumption is:

$$
z \sim \mathcal{N}(\mu, \sigma^2)
$$

To enable backpropagation through random sampling, **VAEs** use the **reparameterization trick**:

$$
z = \mu + \sigma \cdot \epsilon, \quad \epsilon \sim \mathcal{N}(0, 1)
$$

This formulation allows gradients to flow through $\mu$ and $\sigma$ during training.


## Applications of Gaussian Sampling in Generative AI

- **Image Generation:**  
  Latent space vectors sampled from Gaussian priors are decoded into new images using models like Variational Autoencoders (VAEs).

- **Data Augmentation:**  
  Injecting Gaussian noise into input data improves model generalization and robustness.

- **Diffusion Models:**  
  Gaussian noise is added and reversed to generate high-quality images, forming the basis of powerful models like Stable Diffusion.

- **Speech Synthesis:**  
  Gaussian sampling is used in models such as WaveNet to generate realistic audio waveforms.

- **Bayesian Inference:**  
  Gaussian distributions are used to model uncertainty and continuous random variables in probabilistic frameworks.

- **Medical AI:**  
  Gaussian sampling is used to simulate physiological readings or generate synthetic patient data under uncertainty.

# Implementation

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of samples
num_samples = 10000

# Gaussian distribution parameters
mean = 0
std_dev = 1

# Sample from the Gaussian distribution
samples = np.random.normal(loc=mean, scale=std_dev, size=num_samples)

# Plot histogram of samples
plt.figure(figsize=(8, 6))
plt.hist(samples, bins=50, density=True, alpha=0.6, color='skyblue', label='Sampled Data')

# Plot the true Gaussian PDF
x = np.linspace(-4, 4, 1000)
pdf = (1 / (np.sqrt(2 * np.pi) * std_dev)) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
plt.plot(x, pdf, 'r', lw=2, label='True PDF')

# Labels and legend
plt.title("Sampling from a Gaussian Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```
